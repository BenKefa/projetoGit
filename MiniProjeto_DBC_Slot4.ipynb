Converter os graus de Fahrenheit para Celsius

```python
def fahr_to_celsius(temp_fahr):
    '''Converter Fahrenheit em Celsius
    
    Retorna o valor convertido em Celsius'''
    temp_celsius = (temp_fahr - 32) * 5 / 9
    return temp_celsius
```

```python
df ['Temperatura'] = fahr_to_celsius(df['Temperatura'])
df
```

Aqui estou tentando deixar a coluna Temperatura com 2 casas decimais. 


```python
# Essa função é para trucar os números. Exemplo: 1,199999 fica apenas 1,19 e não 2,2. 
def trunc(num, digits):
   sp = str(num).split('.')
   return '.'.join([sp[0], sp[:digits]])
```

```python
df ['Temperatura'] = trunc(df['Temperatura':2])
df
```
